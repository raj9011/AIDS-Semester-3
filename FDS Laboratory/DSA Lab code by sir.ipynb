### Average Score of the class

```python
def Average_Marks():
    total=0
    i=0
    while(i<n):
        if(mrks[i]>0):
            total=total+mrks[i]
            i=i+1
    average=total//n
    print("Average Marks Obtained by Students", average)
    
```

```python
Average_Marks()
```

### Highest score and lowest score of class

```python
def Highest_Lowest(mrks):
    for i in range(0,n):
        for j in range(i+1, n):
            if mrks[i]<mrks[j]:
                #temp=mrks[i]
                #mrks[i]=mrks[j]
                #mrks[j]=temp
                mrks[i],mrks[j]=mrks[j],mrks[i]
                
    print(mrks[0], mrks[n-1])
    
```

```python
Highest_Lowest(mrks)
```

### Count number of students absent 

```python
def Count_Absent(mrks):
    count=0
    for i in range(0,n):
        if mrks[i]<0:
            count=count+1
            print(i)
        continue
    print("Total Absent students", count)
```
